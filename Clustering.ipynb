```python
df = data[['age', 'balance', 'duration', 'campaign']].copy()
df.describe()
```

# ESCALADO

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

# ELBOW METHOD

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
K = range(1, 20)

for k in K:
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(X_scaled)
    inertia.append(model.inertia_)

plt.plot(K, inertia, 'o-')
plt.xlabel('Número de clusters')
plt.ylabel('Inercia')
plt.title('Método del codo')
plt.show()
```

# K-MEANS

```python
kmeans = KMeans(n_clusters=5, random_state=42)
df['cluster'] = kmeans.fit_predict(X_scaled)
```

```python
df['cluster'].value_counts()
```

```python
df.head()
```

# ANALISIS DE COMPONENTES PRINCIPALES (PCA)

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# los transformamos a Dataframe
pca_df = pd.DataFrame(X_pca, columns=['PCA1', 'PCA2'])
pca_df['cluster'] = df['cluster']
```

```python
plt.figure(figsize=(8,6))
plt.scatter(pca_df['PCA1'], pca_df['PCA2'], c=pca_df['cluster'], cmap='viridis')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('Visualización de clusters de clientes')
plt.show()
```

# CLUSTER JERARQUICO

```python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
```

```python
linked = linkage(X_scaled, method='ward')

plt.figure(figsize=(10, 6))
dendrogram(linked, truncate_mode='lastp', p=30, leaf_rotation=90, leaf_font_size=10)
plt.title('Dendrograma de Clustering Jerárquico')
plt.xlabel('Clientes agrupados')
plt.ylabel('Distancia (Ward)')
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster

clusters_h = fcluster(linked, t=4, criterion='maxclust')
```

```python
df['cluster_h'] = clusters_h
df.head()
```

```python
df['cluster_h'].value_counts()
```

```python
pca_df['cluster_h'] = df['cluster_h'].values

plt.figure(figsize=(8,6))
plt.scatter(pca_df['PCA1'], pca_df['PCA2'], 
            c=pca_df['cluster_h'], 
            cmap='viridis', 
            s=40)
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('Clusters Jerárquicos')
plt.show()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.8, min_samples=5)
clusters_db = dbscan.fit_predict(X_scaled)
```

```python
df['cluster_db'] = clusters_db
```

```python
df['cluster_db'].value_counts()
```

```python
pca_df['cluster_db'] = df['cluster_db'].values

plt.figure(figsize=(8,6))
plt.scatter(pca_df['PCA1'], pca_df['PCA2'],
            c=pca_df['cluster_db'],
            cmap='viridis',
            s=40)

plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('Clusters DBSCAN)')
plt.show()
```

```python
print("K-Means:", df['cluster'].nunique(), "clusters")
print("Jerárquico:", df['cluster_h'].nunique(), "clusters")
print("DBSCAN:", df['cluster_db'].nunique(), "clusters")
```

# SILHOUETTE SCORE

```python
from sklearn.metrics import silhouette_score

print("Silhouette K-Means:", silhouette_score(X_scaled, df['cluster']))
print("Silhouette Jerárquico:", silhouette_score(X_scaled, df['cluster_h']))
```

```python
mask = df['cluster_db'] != -1
silhouette_score(X_scaled[mask], df.loc[mask, 'cluster_db'])
```
